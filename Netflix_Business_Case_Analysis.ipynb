```python
# Generally missing value can be treated by using mean, mode or median approach.
# If a small percentage of missing values are there then we may remove the missing values.
# Here in this case around 25% missing values are present.
# Assuming this mising value count is considerably higher and most of the attributes are object type
# so instead of dropping missing values , replacing missing data by following values as per the attribute
# In the column of 'director' missing values will be replace by ' No Director'
# In the column of 'cast' missing values will be replace by ' No Cast'
# In the column of 'country' missing values will be replace by 'Country Unavailable'
# The attributes date_added, rating ,duration may be drop because of a small percentage of missing values.
```

```python
df1.director.fillna("No Director", inplace=True)
df1.cast.fillna("No Cast", inplace=True)
df1.country.fillna("Country Unavailable", inplace=True)
df1.dropna(subset=["date_added", "rating","duration"], inplace=True)
```

```python
max_year = df1['release_year'].max()
start_year = max_year - 30

# Filter for movies released in the last 20-30 years
filtered_df = df1[df1['release_year'].between(start_year,max_year)]
```

```python
# Univariate Analysis:
# Perform a univariate analysis by calculating the count of movies released each year.
# Group by release year and count the number of movies
movies_per_year = filtered_df.groupby('release_year').size().reset_index(name='count')

# Display the resulting DataFrame
movies_per_year
```

```python
plt.figure(figsize=(10, 5))
plt.plot(movies_per_year['release_year'], movies_per_year['count'], marker='o')
plt.title('Number of Movies Released per Year (Last 20-30 Years)')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies Released')
plt.show()
```

- As it is seen from the graph that for last 20 to 30 years the number of movies released increased.
- After 2014 there is sharp increase in the movie relesed.
- For the last 3 years i.e 2019 to 2021 , movies/shows release decreases becauses of the world wide pandamic conditon.

```python
# Plot the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='month_added', y='count', data=month_added_count, palette='viridis')
plt.title('Month-wise Count of Movies Added to Netflix')
plt.xlabel('Month')
plt.ylabel('Number of Movies Added')
plt.show()
```

- From this Analysis it is seen that most of content is added in the month of July and December  considering the best time to make content  avaible on platform because of mid-year, Chrismas festival and New Year celebrtion period in december.
